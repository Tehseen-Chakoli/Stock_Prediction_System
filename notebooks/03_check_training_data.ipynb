```python
import pandas as pd

# Load cleaned training data
df = pd.read_csv("../data/processed/sbi_training_data.csv")
df["Date"] = pd.to_datetime(df["Date"])

print("Shape:", df.shape)

print("\nColumns:")
print(df.columns.tolist())

print("\nTop 5 rows:")
display(df.head())

print("\nMissing values:")
print(df.isnull().sum())
```

```python
print("Target Distribution:\n")
print(df["target_up_7d"].value_counts())

print("\nTarget Percentage:\n")
print(df["target_up_7d"].value_counts(normalize=True))
```

```python
print("Future Return Stats:\n")
print(df["future_return_7d"].describe())
```

```python
import matplotlib.pyplot as plt

# Price trend
plt.figure()
plt.plot(df["Date"], df["Close"])
plt.title("SBI Price Trend")
plt.xlabel("Date")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Future return distribution
plt.figure()
plt.hist(df["future_return_7d"], bins=50)
plt.title("Future 7-Day Return Distribution")
plt.xlabel("Return")
plt.ylabel("Frequency")
plt.show()

# Target distribution
plt.figure()
df["target_up_7d"].value_counts().plot(kind="bar")
plt.title("Target Distribution (Up/Down)")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()

# RSI vs Price
plt.figure()
plt.scatter(df["rsi_14"], df["Close"])
plt.title("RSI vs Price")
plt.xlabel("RSI")
plt.ylabel("Price")
plt.show()
```

```python
import numpy as np

numeric_df = df.select_dtypes(include=[np.number])

corr = numeric_df.corr()

print("Top correlations with target:\n")
print(corr["target_up_7d"].sort_values(ascending=False).head(10))

plt.figure()
plt.imshow(corr, aspect='auto')
plt.title("Correlation Matrix")
plt.colorbar()
plt.show()
```
